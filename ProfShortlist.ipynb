```python
from bs4 import BeautifulSoup
import requests
```

```python
# 1.Fetch the pages (write the website you wish to scrape within parentheses)
base_url="http://dtu.ac.in/modules/facilities/people/faculty/view.php?uname="
result = requests.get(base_url+"skbabu")
# 2.Get the page content
content = result.text
# 3. Create the soup
soup = BeautifulSoup(content, "lxml")
```

```python
print(soup)
```

```python
res=soup.find_all('li')
```

```python
title=[]
jornal=[]
authors=[]
for i in res:
    try:
        temp=i.text.replace('“', '"').replace('”','"')
        temp=temp.split('"')
        authors.append(temp[0])
        title.append(temp[1])
        jornal.append(temp[2])
    except:
        try:
            temp=i.text
            temp=temp.split('“')
            authors.append(temp[0])
            y=temp[1].split('”')
            title.append(y[0])
            jornal.append(y[1])
            
        except:
            x=i.text.split('"')
            print("NEW CASE")
            for j in x:
                print(j)
            continue
```
